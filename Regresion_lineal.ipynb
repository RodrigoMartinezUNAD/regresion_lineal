```python
# Cargue de librerias

import matplotlib.pyplot as plt
from sklearn import linear_model # usando sklear para saber los valores optimos
import seaborn as sns
import numpy as np
import pandas as pd

# Carga del archivo CSV

data = pd.read_csv("C:/Users/Rodrigo/Downloads/Analisis de Datos/regresion_lineal.csv", sep=",")

# Visualizar datos

data
```

```python
# Explorar datos

data.columns
data.info
data.describe
```

```python
#Grafica de dispersión

data.plot.scatter(x="metro",y="precio")
plt.show()
```

```python
#Regresión Lineal

regresion=linear_model.LinearRegression()

#Agregar datos enun array 
meters=data["metro"].values.reshape((-1,1))

#Crear el modelo

modelo=regresion.fit(meters,data["precio"])

print("Interseccion(b)",modelo.intercept_)

#Imprimir la pendiente

print("Pendiente (m)",modelo.coef_)

entrada=[[5],[10],[15],[20],[25],[30],[35]]
predicciones=modelo.predict(entrada)
print(predicciones)
```

```python
data.plot.scatter(x="metro", y="precio", label='Precio por metro')
plt.scatter(entrada, predicciones, color='red')
plt.plot(entrada, predicciones, color='black', label='Línea de regresión')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
